```python
df = df.drop_duplicates()
```

```python
# drop for unwanted data
df = df.drop(columns=["Vehicle Images",'Incomplete_Rides_Reason'])
```

```python
print("Remaining duplicates:", df.duplicated().sum())
```

```python
missing_percent = df.isnull().sum() / len(df) * 100
missing_percent.sort_values(ascending=False)
```

```python
#  Missing handle values 

# numeric aur categorical seprate 
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object','str']).columns
print(num_cols)
print(cat_cols)
```

```python
for col in num_cols:
    percent = df[col].isnull().mean()*100
    
    if percent > 50:
        df=df.drop(columns=col)
        print(col, "drop (>50% missing)")
    
    elif percent > 0:
        df[col]=df[col].fillna(df[col].median())
        
```

```python
for col in cat_cols:
    percent = df[col].isnull().mean()*100
    
    if percent > 100:
        df.drop(columns=col)
        print(col, "droped")
    
    elif percent > 0:
        df[col]=df[col].fillna(df[col].mode()[0])
        print(col, "mode se fill kiya")
```
